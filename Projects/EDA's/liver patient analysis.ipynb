## BAR GRAPH OF TOTAL BILIRUBIN VS DIRECT BILIRUBIN

```python
plt.figure(figsize=(10,10))
ax=plt.axes()
ax.set_facecolor('yellow')

plt.bar(last.Total_Bilirubin,last.Direct_Bilirubin,label='Direct_Bilirubin rise',color='r',width=1)

plt.legend()
plt.xlabel('Total_Bilirubin')
plt.ylabel('Direct_Bilirubin')
plt.title('Total_Bilirubin vs Direct_Bilirubin')
plt.show()
```

# STRIP  PLOT OF TOTAL BILIRUBIN VS DIRECT BILIRUBIN

## HEATMAP

```python
plt.figure(figsize=(7,7))
sns.heatmap(r.corr(),cmap='viridis',vmin=0,vmax=1)
plt.show()
```

## CATPLOT OF TOTAL PROTIEN VS ALBUMIN



```python
v=r.head(30)
sns.catplot(x='Total_Protiens',y='Albumin',hue='Gender',data=v)
plt.show()
```

## DISTPLOT OF TOTAL TOTAL BILIRUBIN

```python
d=r.head(50)
sns.distplot(d.Total_Bilirubin,color='red')
plt.show()
```

## COUNT PLOT OF TOTAL BILIRUBIN


```python
sns.countplot(x='Total_Bilirubin',data=d,palette='colorblind')
plt.show()
```

## COMBINED BAR GRAPH OF TOTAL BILIRUBIN AND DIRECT BILIRUBIN

```python
d.groupby('Age')[['Total_Bilirubin','Direct_Bilirubin']].sum().plot.bar(color=['red','blue'],figsize=(8,8))
plt.ylabel('type:free & paid')
plt.show()
```
